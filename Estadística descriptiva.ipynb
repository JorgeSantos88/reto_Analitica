```python
#Para poder analizar la informaci√≥n, debemos importar la librer√≠a pandas, que nos permite obtener informaci√≥n de 
#manera m√°s sencilla
import pandas as pd
#En la variable data, guardamos la lectura del archivo con los datos
data = pd.read_csv("covid19_tweets.csv")
#Con head, obtenemos los primeros 5 objetos de la base de datos, y podemos visualizar las caracter√≠sticas e 
#informaci√≥n que cada columna contiene
data.head()
```

```python
#Utilizamos el comando tail para obtener los √∫ltimos 5 regsitros en la base de datos
data.tail()
```

```python
#Para obtener los principales par√°metros estad√≠sticos, se utiliza el par√°metro describe
#la tabla arroja √∫nicamente los resultados de variables num√©ricas
data.describe()
```

```python
#Con el comando que incluye np.object, es posible obtener algunos estad√≠sticos, principalmente frecuencias, 
#de algunas otras variables cualitativas
import numpy as np
data.describe(include=np.object)
```

```python
#Hay columnas que no aparecen en ninguna de las dos variables anteriores (ambas variables son booleanas), 
#podemos usar el comando mode para encontrar el valor que m√°s se repite
data['user_verified'].mode()
```

```python
data['is_retweet'].mode()
```

```python
#Con el comando corr, podemos encontrar si hay correlaci√≥n entre las variables
#en este caso, los valores obtenidos son bajos, se concluye que no hay correlaci√≥n significativa entre variables
data.corr()
```

```python
#Para tener una visualizaci√≥n gr√°fica de la distribuci√≥n de algunas de las variables, se usa un boxplot
import matplotlib.pyplot as plt 
col1 = data['user_followers'] 
fig = plt.figure(figsize =(10, 7)) 
plt.boxplot(col1)
plt.show() 
```

```python
col1 = data['user_friends'] 
fig = plt.figure(figsize =(10, 7)) 
plt.boxplot(col1)
plt.show() 
```
